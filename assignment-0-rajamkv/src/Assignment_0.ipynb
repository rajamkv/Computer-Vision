```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

### *2. Load the images using OpenCV.*

```python
img_1 = cv2.imread('../images/image_1.jpg')
img_2 = cv2.imread('../images/image_2.jpg')
img_3 = cv2.imread('../images/image_3.jpg')
img_4 = cv2.imread('../images/image_4.jpg')

# The imread reads images in BGR format, So , changing them  to RGB

img_1 = cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB)
img_2 = cv2.cvtColor(img_2, cv2.COLOR_BGR2RGB)
img_3 = cv2.cvtColor(img_3, cv2.COLOR_BGR2RGB)
img_4 = cv2.cvtColor(img_4, cv2.COLOR_BGR2RGB)
```

### *3. Display the images using matplotlib (make a 2x2 grid using subplots)*

```python
fig = plt.figure(figsize=[10,10])
ax1 = fig.add_subplot(2,2,1)
ax1.set_title("Image - 1")
ax1.imshow(img_1)

ax2 = fig.add_subplot(2,2,2)
ax1.set_title("Image - 2")
ax2.imshow(img_2)

ax3 = fig.add_subplot(2,2,3)
ax1.set_title("Image - 3")
ax3.imshow(img_3)

ax4 = fig.add_subplot(2,2,4)
ax1.set_title("Image - 4")
ax4.imshow(img_4)

plt.show()
```

### *4. Image to Grayscale*

```python
dim = img_1.shape
Gim = np.zeros((dim[0],dim[1]))

Gim=0.2989*img_1[:,:,0]+0.5870*img_1[:,:,1]+0.1140*img_1[:,:,2]


fig = plt.figure(figsize=[15,15])

ax1 = fig.add_subplot(1,2,1)
ax1.imshow(img_1)

ax1 = fig.add_subplot(1,2,2)
ax1.imshow(Gim,cmap='gray')

plt.show()
```
